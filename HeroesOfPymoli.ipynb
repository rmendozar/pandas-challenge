## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

Dict = df.groupby('Gender').apply(lambda x: x.SN.nunique()).to_dict()

Demo = pd.DataFrame()
Demo['Gender'] = Dict.keys()
Demo['Players'] = Dict.values()
Demo['%'] = Demo.Players / Demo.Players.sum()
Demo = Demo.sort_values('Players',ascending=False)
Demo.reset_index(drop=True, inplace=True)

format_dict = {"Players":'{:,.0f}',"%": '{:.1%}'}
style = Demo.style.format(format_dict)

from IPython.display import display
display(style)

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
def f(x):
    d = {}
    d['Purchase count'] = x['Purchase ID'].count()
    d['Price avg'] = x['Price'].mean()
    d['Total Price'] = x['Price'].sum()
    #d['Players'] = (lambda x: x.SN.nunique()))
    d['Purchase per player'] = (x['Price'].sum() / x.SN.nunique()
    return pd.Series(d, index=['Purchase count', 'Price avd', 'Total Price', 'Purchase per Player'])
```

```python
Gender = df.groupby('Gender').agg({'Purchase ID':'count', 
                         'Price':['mean','sum'],
                          'SN':lambda x: x.nunique()
                         })
Gender = Gender.rename(columns={'SN': 'Players','<lambda>':'Total'})
Gender['Purchase avg'] = Gender[('Price', 'sum')] / Gender[('Players', 'Total')]
Gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 18, 25, 30, 32, 40]

# Create the names for the four bins
group_names = ["0 to 18", "19 to 25", "26 to 32", "33 to 40", "+40"]
df["age_summary"] = pd.cut(df["Age"], bins, labels=group_names)
# Create a GroupBy object based upon "View Group"

Age = df.groupby(['Gender', 'age_summary']).agg({'Purchase ID':'count', 
                         'Price':['mean','sum'],
                          'SN':lambda x: x.nunique()
                         })

Age = Age.rename(columns={'SN': 'Players','<lambda>':'Total'})
Age['Purchase avg'] = Age[('Price', 'sum')] / Age[('Players', 'Total')]
Age.columns
```
